### What should we clean up?

- [x] Get rid of bullet points in Used Names
- [x] Split height/weight
- [x] Parse out dates from `Born` & `Died` column
- [x] Parse out city, region, and country from `Born` column
- [x] Get rid of extra columns

```python
df = bios.copy()
```

```python
df['name'] = df['Used name'].str.replace("•", " ")  # we used str for string
```

```python
df.head()
```

```python
# Split the Measurements column into height & weight columns
df[['height_cm', 'weight_kg']] = df['Measurements'].str.split('/', expand=True)  # expand: make height and weight in different columns
```

```python
# Remove any non-numeric characters (like "cm", "kg", or spaces) using regex
# \D+ matches one or more non-digit characters and replaces them with nothing
#df['height_cm'] = pd.to_numeric(df['height_cm'].str.replace(r'\D+', '', regex=True), errors='coerce')

# pd.to_numeric converts the cleaned strings into numeric values
# errors='coerce' ensures invalid values are turned into NaN instead of raising errors
df['height_cm'] = pd.to_numeric(df['height_cm'].str.strip(' cm'), errors='coerce')
df['weight_kg'] = pd.to_numeric(df['weight_kg'].str.strip(' kg'), errors='coerce')
```

```python
df.sort_values('height_cm', ascending=False)
```

```python
# Parse out dates from 'Born' and 'Died' columns

# a full date like "12 March 1980" (day + month (d{4})+ year d{4}) or a year
date_pattern = r'(\d+ \w+ \d{4}|\d{4})'

# Extract the date (day, month, year) or year only from the "Born" column
df['born_date'] = df['Born'].str.extract(date_pattern) # 12 December 1886	

# Extract only the 4-digit (d{4}) year from the "Born" column
df['born_year'] = df['Born'].str.extract(r'(\d{4})')  # 1886

# Convert the "born_date" column into a proper datetime format
# format="mixed" allows mixed formats (just year or full date)
# errors='coerce' turns invalid values into NaT (Not a Time)
#df['born_date'] = pd.to_datetime(df['born_date'], format="mixed", errors='coerce')  # example: 2 10 1960

# Do the same extraction for the "Died" column
df['died_date'] = df['Died'].str.extract(date_pattern)

# Convert the "died_date" into proper datetime
df['died_date'] = pd.to_datetime(df['died_date'], format='mixed', errors='coerce') # convert month to number not string

df['died_date'].dt.strftime("%d-%B-%Y")  # put - between day and month and year

    # Preview the updated DataFrame
df.head()
```

```python
# Get city, region, and country from Born column
location_pattern = r'in ([\w\s()-]+), ([\w\s-]+) \((\w+)\)'
df['Born'].str.extract(location_pattern)
```

```python
df[['born_city','born_region','born_country']] = df['Born'].str.extract(location_pattern, expand=True)
```

```python
#counts how many athletes were born in each city
import pandas as pd
pd.set_option('display.max_colwidth', None)

df.value_counts('born_city',ascending=False).head(25)
```

```python
# Remove any unneccessary columns
df.columns
```

```python
columns_to_keep = ['athlete_id', 'name', 'born_date', 'born_city', 'born_region',
                   'born_country', 'NOC', 'height_cm', 'weight_kg', 'died_date']

df_clean = df[columns_to_keep]
```

```python
df_clean.to_csv('bios_clean.csv', index=False)
```
